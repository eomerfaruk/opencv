```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
img = cv2.imread("3.1 h_line.png (1).png")
```

```python
gray =  cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
```

```python
edges = cv2.Canny(gray , 75,150)
```

```python
plt.imshow(img)
```

```python
plt.imshow(edges)
```

```python
lines = cv2.HoughLinesP(edges,1,np.pi/180,50)
```

```python
for line in lines:
    print(line) # iç içe liste yapısı döndürür
```

```python
for line in lines:
    print(line[0])
```

```python
for line in lines:
    x1,y1,x2,y2 =line[0] # cizgilerin başlangıc ve bitiş noktalarına erişildi
    cv2.line(img,(x1,y2),(x2,y2) , (0,0,255),2)
    
plt.imshow(img)
     
```
