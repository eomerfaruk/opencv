```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
img1 = cv2.imread("5.1 coins.jpg (1).jpg")
```

```python
plt.imshow(img1)
```

```python
img2 = cv2.imread("5.2 balls.jpg (1).jpg")
```

```python
gray_coin = cv2.cvtColor(img1 ,cv2.COLOR_BGR2GRAY )
gray_ball = cv2.cvtColor(img2 ,cv2.COLOR_BGR2GRAY )
```

```python
gray_coin_blur = cv2.medianBlur(gray_coin , 5)
gray_ball_blur = cv2.medianBlur(gray_ball , 5)
```

```python
circles = cv2.HoughCircles(gray_coin_blur ,cv2.HOUGH_GRADIENT,1,img1.shape[0] / 64,param1 = 200,param2 = 10 , minRadius=5,maxRadius=30) # grntdeki daireleri bulur
```

```python
print(circles)
```

```python
if circles is not None:
    circles = np.uint16(np.around(circles))  # circle iinde tutulan deerlr yuvarlanr ilenebilir olmas iin
                                            
```

```python
circles
```

```python
circles.shape
```

```python
for i in circles[0,:]:
    cv2.circle(img1 ,(i[0],i[1]) ,i[2] ,(0,255,0),2)
```
