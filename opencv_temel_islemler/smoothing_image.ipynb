```python
import cv2 
import numpy as np 
from matplotlib  import pyplot as plt
```

```python
img_filter = cv2.imread("8.2 filter.png (1).png")
plt.imshow(img_filter)
```

```python
img_bileteral = cv2.imread("8.4 bilateral.png (1).png")
plt.imshow(img_bileteral)
```

```python
img_median = cv2.imread("8.1 median.png (1).png")
plt.imshow(img_median)
```

```python
blur = cv2.blur(img_filter ,(5,5))
plt.imshow(blur)
```

```python
gaussian_blur = cv2.GaussianBlur(img_filter , (5,5),cv2.BORDER_DEFAULT)
```

```python
plt.imshow(gaussian_blur)
```

```python
median_blur =cv2.medianBlur(img_median ,11)
plt.imshow(median_blur)
```

```python
bileteral_blur = cv2.bilateralFilter(img_bileteral , 9,75,75)
plt.imshow(bileteral_blur)
```
