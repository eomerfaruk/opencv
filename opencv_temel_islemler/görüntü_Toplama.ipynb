```python
import cv2 
from matplotlib import pyplot as plt 
import numpy as np 
```

```python
img = cv2.imread("14.1 smile.jpg.jpg")
plt.imshow(img)
```

```python
circle = np.zeros((512,512 , 3), np.uint8)+255
```

```python
cv2.circle(circle,(256,256) , 60 , (255,0,0) , -1) # circle grntsne  daire cizildi 
```

```python
plt.imshow(circle)
```

```python
rectangle = np.zeros((512,512 , 3), np.uint8)+255
```

```python
cv2.rectangle(rectangle ,(150,150) ,(350,350) , (0,255,0 ),-1)
```

```python
plt.imshow(rectangle)
```
