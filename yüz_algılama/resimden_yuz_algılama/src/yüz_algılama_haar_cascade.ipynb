```python
face_cascade = cv2.CascadeClassifier("4.1 frontalface.xml (1).xml") #Kullanacak cascade dosyasını çalışmaya  dahil etmek
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray
```

```python
plt.imshow(gray)
```

```python
output_img = img
```

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 7)#cacade dosyasını kullanarak resim üstündeki yüzlerin kordinatları tespit edilir
```

```python
# faces değişkeninee tutulan kordinatlarıkullanarak yüzleri dikörtgen içine cizilir
for (x,y,w,h) in faces:
    cv2.rectangle(output_img , (x,y),(x+w,y+h) ,(0,0,255) ,2)
```

```python
plt.imshow(output_img)
```

```python
output_dir = "opencv/yüz_algılama/output" #kasöre yüklenenek dosya yolu
os.makedirs(output_dir, exist_ok=True) # belirtilen dizini (klasörü) oluşturmak için kullanılır.
output_path = os.path.join(output_dir, "yüz_algılanan_img.png")# Kayıt yolu ve dosya adı
output  =cv2.imwrite(output_path,output_img) # dosya yoluna görüntüyü ekler
```
