```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
img =  cv2.imread("5.1 starwars.jpg (1).jpg")
```

```python
template = cv2.imread("5.2 starwars2.jpg (1).jpg")
```

```python
gray_img = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
gray_template = cv2.cvtColor(template , cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img)
```

```python
plt.imshow(gray_template)
```

```python
result = cv2.matchTemplate(gray_img , gray_template, cv2.TM_CCOEFF_NORMED) # şablon eşleştirme
```

```python
result # cıktıda iki görüntüdeki piksellerin  -1 ile 1 arasındaki kolerasyonu verir
```

```python
location =  np.where(result >= 0.7) # result değeri 0.7 ve üstü olan değerlerin kordinatlari
```

```python
location #  tuple içinde iki liste döndürür
```

```python
w,h = gray_template.shape[::-1]
```

```python
for p in zip(*location):
    print(p)
```

```python
for point in zip(*location[::-1]):
    print(point)
    print(point[0])
```

```python
for point in zip(*location[::-1]):
    cv2.rectangle(img  , point ,(point[0] +w , point[1]+h) , (0,255,0),2)
```
