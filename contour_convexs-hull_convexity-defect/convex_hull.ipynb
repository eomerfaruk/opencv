```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
img = cv2.imread("8.1 map.jpg (1).jpg")
```

```python
gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
blur = cv2.blur(gray ,(3,3))
_ , thresh = cv2.threshold(blur , 40,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh)
```

```python
contours , hierarchy = cv2.findContours(thresh , cv2.RETR_TREE ,cv2.CHAIN_APPROX_SIMPLE)
```

```python
hull = []
for i in range(len(contours)):
    hull.append(cv2.convexHull(contours[i] ,False) ) # dış bükey kısımları bulur
    
```

```python
bg = np.zeros((thresh.shape[0],thresh.shape[1],3),np.uint8 )    
```

```python
for i in range(len(contours)):
    cv2.drawContours(bg , contours,i,(0,0,255),3,8,hierarchy) #i : belirli indexi  gösterir 
    cv2.drawContours(bg , hull,i,(0,0,255),1,8) 
```

```python
plt.imshow(bg)
```
