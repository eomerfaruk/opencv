```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
img = cv2.imread("15.2 contour.png.png")
img
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
```

```python
_ , threshold = cv2.threshold(gray , 127 ,255 , cv2.THRESH_BINARY)
```

```python
M = cv2.moments(threshold)  # görüntünün özelliklerini gösterir
M
```

```python
# geometri merkezinin koordinatları
X = int(M["m10"]/M["m00"])
Y = int(M["m01"]/M["m00"])
```

```python
cv2.circle(img , (X,Y),5, (0,0,255) , -1)
```

```python
plt.imshow(img)
```
