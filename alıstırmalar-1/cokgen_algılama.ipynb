```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
img = cv2.imread("1.1 polygons.png (1).png")
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
```

```python
_,thresh = cv2.threshold(gray , 240 ,255 , cv2.THRESH_BINARY)
```

```python
contour , _ = cv2.findContours(thresh , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE)
```

```python
img1 = img
```

```python
cv2.drawContours(img1,contour,-1,(0,255,0),2)
```

```python
plt.imshow(img1)
```

```python
for cnt in contour:
    epsilon = 0.01*cv2.arcLength(cnt,True) # Konturun evresine bal olarak hassasiyet belirlenir
    approx = cv2.approxPolyDP(cnt , epsilon , True) # Kontur yaklak hale getirilir
    
    cv2.drawContours(img ,[approx] ,0,(0,255,0),2)
    
    
```

```python
plt.imshow(img)
```
